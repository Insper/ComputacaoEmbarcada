```python
import neurokit2 as nk  # Load the package
import numpy as np
```

```python
simulated_ecg = nk.ecg_simulate(duration=2, sampling_rate=1000, heart_rate=80)

data = simulated_ecg[330:1250]

delta = data.max() + abs(data.min())
scale = 3.0 / delta
data = (data + abs(data.min())) * scale

data = data * 4095/3.0

nk.signal_plot(data, sampling_rate=1000)  # Visualize the signal

d = np.concatenate([data, data])
nk.signal_plot(d, sampling_rate=1000)  # Visualize the signal
```
